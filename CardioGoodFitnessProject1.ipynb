# Background
This dataset is for customers who purchased Treadmill product(s) from a retail store called Cardio Good Fitness.

# Objective 
* Come up with a customer profile (characteristics of a customer) for each Treadmill product
* Perform uni-variate and multi-variate analyses
* Generate a set of insights and recommendations that will help the company in targeting new customers

# Key Questions to be Answered 
* Are there any differences between customers of each product?
* Is there any relationship patterns that can be observed between the different attributes of customers?


# Dataset 
* Product - the model no. of the treadmill
* Age - in no of years, of the customer
* Gender - of the customer
* Education - in no. of years, of the customer
* Marital Status - of the customer
* Usage - Avg. # times the customer wants to use the treadmill every week
* Fitness - Self rated fitness score of the customer (5 - very fit, 1 - very unfit)
* Income - of the customer
* Miles- expected to run

### Another way to visualize distribution of various numerical data using histogram plot with mean and median on the plots

```python
# to visualize distribution of various important customer profile data (numerical data) using histogram plot of 
# Matplotlib package and plot the mean and median on the plot 
# mean = color code is red solid line
# median = color code is blue dotted line

cust_profile = ['Age', 'Education', 'Usage', 'Miles','Fitness','Income']
#list(enumerate(colors))
#  plot Numerical Data
a = 3  # number of rows
b = 2  # number of columns
c = 1  # initialize plot counter

fig = plt.figure(figsize=(14,22))
for i in cust_profile:
    plt.subplot(a, b, c)
    plt.title('{} (hist), subplot: {}{}{}'.format(i, a, b, c),fontsize=12,fontweight='bold')
    plt.hist(cardiofit[i],color='coral') 
    plt.axvline(cardiofit[i].mean(), color='r', linewidth=2)
    plt.axvline(cardiofit[i].median(), color='b', linestyle='dashed', linewidth=2)       
    c = c + 1
    continue

plt.show()





```

### Observations
* Mean customer age is 28 yeras old and Median is 26 years old. Most of the customers are < 35 years old. Customers drop > 35 years old.
* Income: Mean income is 52K USD per annum larger than median income 50K USD per annum. Most of the customers fall in first 4 bins, that is in lower pay range < 60K USD per annum 
* Most of the customers expect to run 100 miles per week, mean and median is pretty close. Very few customers expect to run > 150 miles per week. 
* There is not much variance of data with Education, Fitness and Usage 
* Mean and Median self rated fitness scale is pretty close. Mean is 3.3 and Median is 3.0. Most of the customers fall in 3.0 bin
* Mean expected usage of Treadmill is 3.4 and Median expected usage of Treadmill is 3.0. Most of the customers fall in the 3 days per week bin
* Mean education is 16 years, close to the median. Most of the customers fall in the 16 years bin.
